```python
df=netflix.copy()
```

```python
df=df.dropna()
df.shape
```

```python
df.head()
```

```python
import pandas as pd

df["date_added"] = pd.to_datetime(df['date_added'], errors='coerce')

# Check for any NaT (Not a Time) values after parsing
invalid_dates = df['date_added'].isna()
if invalid_dates.any():
    print(f"Warning: {invalid_dates.sum()} rows have invalid date format.")

df['day_added'] = df['date_added'].dt.day
df['year_added'] = df['date_added'].dt.year

# Fill NaN values with a placeholder (you can replace -1 with any suitable placeholder)
df['year_added'] = df['year_added'].fillna(-1).astype(int)
df['day_added'] = df['day_added'].fillna(-1).astype(int)

```

```python
df.head()
```

```python
sns.countplot(x="type", data=df, palette="deep")
plt.show()
```

```python
sns.countplot(netflix['rating'])
sns.countplot(netflix['rating']).set_xticklabels(sns.countplot(netflix['rating']).get_xticklabels(), rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(13,13)
plt.title('Rating')
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='rating',hue='type',data=netflix)
plt.title('Relation between Type and Rating')
plt.show()
```

```python
labels = ['Movie', 'TV show']
size = netflix['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```

```python
netflix['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.show()
```
